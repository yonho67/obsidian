```python
from pathlib import Path
import re

# 1. 이름 변경 작업을 진행할 대상 디렉토리 설정 
# (현재 디렉토리 기준이면 Path('.') / 절대 경로면 Path(r"/mnt/g/..."))
TARGET_DIR = Path('.')

# 2. 매칭할 패턴 정의: essay_v 로 시작하고 뒤에 숫자가 붙는 폴더
# 예: essay_v80, essay_v95m 등
folder_pattern = re.compile(r'^essay_v(\d+)(.*)$')

print("🔄 버전 폴더 이름 변경 시작...")
print("-" * 50)

# 대상 디렉토리 안의 모든 하위 아이템 검사
for folder in TARGET_DIR.iterdir():
    # 폴더(디렉토리)인 경우에만 진행
    if folder.is_dir():
        match = folder_pattern.match(folder.name)
        if match:
            # 기존 숫자와 그 뒤에 붙은 문자(예: m, g 등) 추출
            current_version = int(match.group(1))
            suffix = match.group(2)
            
            # 버전 번호에서 70 빼기
            new_version = current_version - 70
            
            # 새 폴더 이름 생성
            new_name = f"essay_v{new_version}{suffix}"
            new_path = folder.with_name(new_name)
            
            # 이름 변경 실행 및 로그 출력
            print(f"📄 [변경] {folder.name} ➡️ {new_name}")
            folder.rename(new_path)

print("-" * 50)
print("✅ 모든 폴더 이름 변경이 완료되었습니다!")
```
