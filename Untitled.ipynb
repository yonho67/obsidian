```python
 🥇 Kiwipiepy — 가장 권장
  - 순수 pip 설치, Java JDK 불필요 (Windows 친화적)
  - C++ 백엔드라 속도 빠름, 정확도 높음
  - 활발히 유지보수됨
  - pip install kiwipiepy

  다른 선택지:
  - KoNLPy + Okt: 가장 유명하나 Java 필요
  - KoNLPy + Mecab: 가장 빠르나 Windows 설치 까다로움
  - KoNLPy + Komoran: 순수 Java, 설치 쉬움

  실행 스크립트 (Kiwipiepy 기준)

  # count_morphs.py
  from pathlib import Path
  from collections import Counter
  import csv
  from kiwipiepy import Kiwi

  ROOT = Path(r"G:\내 드라이브\my_vault\essay_v99")
  OUT = Path(r"E:\projects\book-rag\v99_빈도분석.csv")

  # 세고 싶은 품사 (Kiwi 태그):
  # NNG=일반명사, NNP=고유명사, VV=동사, VA=형용사, MAG=일반부사
  TARGET_TAGS = {"NNG", "NNP", "VV", "VA"}

  # 제외 파일 (목차·집필 원칙 등)
  EXCLUDE_PATTERNS = ["집필 원칙", "목차", "단원별", "전체", "수식", "wordcount"]

  kiwi = Kiwi()
  counter = Counter()  # (lemma, pos) -> count
  per_file = {}        # filename -> Counter

  for md in ROOT.rglob("*.md"):
      if any(pat in md.name for pat in EXCLUDE_PATTERNS):
          continue
      text = md.read_text(encoding="utf-8")
      file_counter = Counter()
      for token in kiwi.tokenize(text):
          if token.tag in TARGET_TAGS:
              # token.form = 어간 (동사 "봤다" → "보")
              # token.tag = 품사
              key = (token.form, token.tag)
              counter[key] += 1
              file_counter[key] += 1
      per_file[md.relative_to(ROOT).as_posix()] = file_counter

  # 결과 저장
  with OUT.open("w", encoding="utf-8-sig", newline="") as f:
      w = csv.writer(f)
      w.writerow(["어휘", "품사", "전체 빈도"])
      for (form, tag), cnt in counter.most_common():
          w.writerow([form, tag, cnt])

  print(f"완료: {OUT}")
  print(f"고유 형태소 수: {len(counter)}")
  print(f"총 토큰 수: {sum(counter.values())}")

  사용법

  # 1) 의존성 설치
  uv add kiwipiepy   # 또는 pip install kiwipiepy

  # 2) 실행
  uv run python count_morphs.py

  응용 예시

  특정 단어만 추적하고 싶다면 마지막 부분만 바꾸면 됩니다:

  # 특정 어휘 목록만 추적
  WATCH = {"세계", "우주", "자연", "현실", "실재", "보다", "있다"}
  for (form, tag), cnt in counter.items():
      if form in WATCH:
          print(f"{form}({tag}): {cnt}")

  또는 챕터별 표를 만들고 싶으면 per_file을 마크다운 표로 출력하면 됩니다.

  원하시면 이 스크립트를 E:\projects\book-rag\src\ 아래에 만들고 실행해 결과 CSV까지 생성해 드리겠습니다. 명령만 주시면
  진행하겠습니다.

✻ Sautéed for 44s
```
